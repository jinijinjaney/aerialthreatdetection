```python
!pip install -q ultralytics gdown roboflow opencv-python matplotlib seaborn

import os
import cv2
import gdown
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from ultralytics import YOLO
from collections import Counter

np.random.seed(42)

print("Downloading and preparing dataset")
gdown.download(id='1li09Zre8jKX9MIZXdAFpxDsor369Ndgc', output='AerialDataset.zip', quiet=False)

with zipfile.ZipFile("AerialDataset.zip", 'r') as zip_ref:
    zip_ref.extractall("AerialDataset")

base_path = "/content/AerialDataset/AerialDataset"

def verify_and_fix_labels(base_path):
    """Enhanced label verification and correction"""
    label_dirs = [
        f"{base_path}/train/labels",
        f"{base_path}/valid/labels",
        f"{base_path}/test/labels"
    ]

    for label_dir in label_dirs:
        base_dir = Path(label_dir)
        print(f"Verifying and fixing labels in: {base_dir}")

        for label_file in base_dir.glob("*.txt"):
            with open(label_file, 'r') as f:
                lines = f.readlines()

            new_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) < 5:
                    continue

                cls = parts[0]
                coords = parts[1:5]

                try:
                    cls = int(cls)
                    if cls not in [0, 1]:
                        continue

                    coords = [float(x) for x in coords]
                    if any(c < 0 or c > 1 for c in coords):
                        continue

                    new_lines.append(f"{cls} {' '.join(parts[1:])}\n")
                except ValueError:
                    continue

            with open(label_file, 'w') as f:
                f.writelines(new_lines)

        print(f"Completed: {base_dir}")

verify_and_fix_labels(base_path)

def analyze_class_distribution(base_path):
    """Comprehensive class distribution analysis with visualization"""
    splits = ["train", "valid", "test"]
    class_names = {0: "Civilian", 1: "Soldier"}

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    for i, split in enumerate(splits):
        label_dir = f"{base_path}/{split}/labels"
        if not os.path.exists(label_dir):
            continue

        counts = Counter()
        for label_file in Path(label_dir).glob("*.txt"):
            with open(label_file) as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) >= 1:
                        cls = int(parts[0])
                        counts[cls] += 1

        ax = axes[i]
        sns.barplot(x=[class_names[c] for c in counts.keys()],
                   y=list(counts.values()), ax=ax)
        ax.set_title(f"{split.capitalize()} Set Class Distribution")
        ax.set_ylabel("Count")
        ax.set_xlabel("Class")

        for p in ax.patches:
            ax.annotate(f'{int(p.get_height())}',
                       (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center',
                       xytext=(0, 5), textcoords='offset points')

    plt.tight_layout()
    plt.savefig("class_distribution.png")
    plt.show()
    print("Class distribution analysis completed")

analyze_class_distribution(base_path)

print("Starting model training...")
model = YOLO('yolov8m.pt')

training_params = {
    'data': f"{base_path}/data.yaml",
    'epochs': 150,
    'imgsz': 640,
    'batch': 16,
    'patience': 20,
    'device': '0',
    'workers': 4,
    'optimizer': 'AdamW',
    'lr0': 0.001,
    'lrf': 0.01,
    'momentum': 0.937,
    'weight_decay': 0.0005,
    'warmup_epochs': 5.0,
    'warmup_momentum': 0.8,
    'warmup_bias_lr': 0.1,
    'box': 7.5,
    'cls': 1.0,
    'dfl': 1.5,
    'fliplr': 0.5,
    'mosaic': 1.0,
    'mixup': 0.2,
    'copy_paste': 0.2,
    'hsv_h': 0.015,
    'hsv_s': 0.7,
    'hsv_v': 0.4,
    'degrees': 15.0,
    'translate': 0.2,
    'scale': 0.5,
    'shear': 5.0,
    'perspective': 0.001,
    'name': 'aerial-threat-detection-v3',
    'cos_lr': True,
    'label_smoothing': 0.1,
    'nbs': 64,
    'val': True,
    'save_period': 10,
    'seed': 42
}

results = model.train(**training_params)
print("TRAINING COMPLETED")
```

```python
from pathlib import Path
import matplotlib.pyplot as plt
import cv2
import time
from IPython.display import display, Image
from google.colab import files
from ultralytics import YOLO  # Make sure you have this import

# Load the model explicitly from best.pt
model_path = '/content/runs/detect/aerial-threat-detection-v3/weights/best.pt'
print(f"Loading model from {model_path}...")
model = YOLO(model_path)

# === Step 1: Evaluate the model ===
print("Evaluating model performance...")
metrics = model.val(
    data=f"{base_path}/data.yaml",
    split='test',
    conf=0.25,
    iou=0.6,
    plots=True,
    save_json=True,
    save_hybrid=True
)

print("\nModel Evaluation Metrics:")
print(f"mAP@0.5: {metrics.box.map50:.4f}")
print(f"mAP@0.5:0.95: {metrics.box.map:.4f}")
print(f"Precision: {metrics.box.p.mean():.4f}")
print(f"Recall: {metrics.box.r.mean():.4f}")

# === Step 2: Download PR Curve ===
files.download('/content/runs/detect/aerial-threat-detection-v37/PR_curve.png')

# === Step 3: Prediction + Visualization ===
def enhanced_predict_and_visualize(model, base_path):
    test_images = list(Path(f"{base_path}/test/images").glob("*.jpg"))[:5]

    for img_path in test_images:
        start_time = time.time()
        results = model.predict(source=str(img_path), save=True, conf=0.3)
        inference_time = time.time() - start_time

        pred_dir = results[0].save_dir if results else None
        pred_img_path = Path(pred_dir) / img_path.name

        if pred_img_path.exists():
            detected_img = cv2.imread(str(pred_img_path))
            detected_img = cv2.cvtColor(detected_img, cv2.COLOR_BGR2RGB)

            plt.figure(figsize=(10, 10))
            plt.imshow(detected_img)
            plt.title(f"Detection Results (Inference Time: {inference_time:.3f}s)")
            plt.axis('off')
            plt.show()
        else:
            print(f"[!] Warning: Predicted image not found for {img_path.name}")

        for result in results:
            print(f"\nImage: {img_path.name}")
            print(f"Inference Time: {inference_time:.3f}s")
            for box in result.boxes:
                cls = int(box.cls[0])
                conf = float(box.conf[0])
                print(f"Detected: {'Soldier' if cls == 1 else 'Civilian'} with confidence {conf:.2f}")

enhanced_predict_and_visualize(model, base_path)

# === Step 4: Export model ===
print("Exporting model...")
model.export(format='onnx', dynamic=True, simplify=True)

# === Step 5: Download files ===
files.download('/content/runs/detect/aerial-threat-detection-v3/weights/best.pt')  # best.pt download
files.download('/content/runs/detect/aerial-threat-detection-v3/weights/best.onnx')
files.download('/content/aerial_threat_detection.pt')

# Safe attempts for training-only outputs
try:
    files.download('/content/runs/detect/aerial-threat-detection-v3/confusion_matrix.png')
except:
    print("[!] confusion_matrix.png not found")

try:
    files.download('/content/runs/detect/aerial-threat-detection-v3/results.png')
except:
    print("[!] results.png not found")

try:
    files.download('/content/class_distribution.png')
except:
    print("[!] class_distribution.png not found")
```
